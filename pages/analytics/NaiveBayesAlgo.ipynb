```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_auc_score, f1_score, precision_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import ConfusionMatrixDisplay

# Split the data into train and test sets
X = df.drop(columns=['Pre-Diagnosis'])
Y = df['Pre-Diagnosis']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)

# Train the Gaussian Naive Bayes model
model = GaussianNB()
model.fit(x_train, y_train)

# Generate predictions on the test set
y_pred = model.predict(x_test)

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate AUC score for each class
auc_scores = []
for i in range(cm.shape[0]):
    y_true_class = np.where(y_test == class_labels[i], 1, 0)
    y_pred_class = np.where(y_pred == class_labels[i], 1, 0)
    auc_scores.append(roc_auc_score(y_true_class, y_pred_class))

# Calculate average AUC score
auc = np.mean(auc_scores)

# Calculate F1 score
f1 = f1_score(y_test, y_pred, average='macro')

# Calculate precision
precision = precision_score(y_test, y_pred, average='macro')

# Display the confusion matrix with class labels
class_labels = df['Pre-Diagnosis'].unique()
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)

# Increase the size of the confusion matrix plot
fig, ax = plt.subplots(figsize=(24, 13))  # Adjust the figsize as desired

disp.plot(ax=ax)


# Add text box for TP, TN, FP, FN, AUC, F1, and Precision
textstr = f'True Positives (TP) = {cm[1, 1]}\n' \
          f'True Negatives (TN) = {cm[0, 0]}\n' \
          f'False Positives (FP) = {cm[0, 1]}\n' \
          f'False Negatives (FN) = {cm[1, 0]}\n' \
          f'AUC = {auc:.4f}\n' \
          f'F1 Score = {f1:.4f}\n' \
          f'Precision = {precision:.4f}'

props = dict(boxstyle='round', facecolor='white', alpha=0.5)
ax.text(0.6, 0.3, textstr, transform=ax.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)

plt.xticks(fontsize=6)
plt.yticks(fontsize=8)

plt.show()
```
